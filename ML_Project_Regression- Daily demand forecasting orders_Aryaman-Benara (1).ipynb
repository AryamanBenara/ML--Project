```python
#Removing the outliers
def outliers(df, feature):
    Q1= df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    return upper_limit, lower_limit

upper, lower = outliers(df, "Banking orders (3)")
print("Upper whisker: ", upper)
print("Lower Whisker: ", lower)
```

```python
#Calculating the dependent and independent variables
from sklearn.model_selection import train_test_split
y = df['Target (Total orders)']
X = df.loc[:, df.columns != 'Target (Total orders)']
print(X.shape)
```

```python
#Assigning the test and train variables
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python
#Printing the test and train variables
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Mentioning the first few rows of the train variable
X_train.head()
```

```python
#Mentioning the first few rows of the train variable
y_train.head()
```

```python
#Building the model
# Linear Regression !
from sklearn import linear_model
from sklearn.metrics import r2_score
linear_model = linear_model.LinearRegression()
lr_model = linear_model.fit(X_train, y_train)
Y_pred_lr =linear_model.predict(X_train)
y_pred_test_lr = linear_model.predict(X_test)
```

```python
#Calculating the accuracy of the model built for daily demand forecasting orders
residuals = y_test - y_pred_test_lr
print(r2_score(y_test , y_pred_test_lr))
```
